```python
#Task 1


class Student:
    id=0
    def __init__(self,name,dept,age,cgpa):
        self.name=name
        self.dept=dept
        self.age=age
        self.cgpa=cgpa
        Student.id+=1

    def get_details(self):
        print('ID:',Student.id)
        print('Name:',self.name)
        print('Department:',self.dept)
        print('Age:',self.age)
        print('CGPA:',self.cgpa)


    @classmethod
    def from_String(self,stu_info):
        x=stu_info.split('-')
        return Student(x[0],x[1],x[2],x[3])

s1 = Student("Samin", "CSE", 21, 3.91)
s1.get_details()
print("-----------------------")
s2 = Student("Fahim", "ECE", 21, 3.85)
s2.get_details()
print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.get_details()
print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")
s4.get_details()

#subtask 5
print('difference between a class variable and an instance variable is instance variables are differnt for each objects but\n'
      'class variables are shared between all objects within a class.')

#subtask 6
print('difference between an instance method and class method: instance method used only for object\n'
      'and class method used for accessing the class properties')
```

```python
#Task 2


class Assassin:
    no=0
    def __init__(self,name,success_rate):
        self.name=name
        self.success_rate=success_rate
        Assassin.no+=1

    @classmethod
    def failureRate(cls,name,failure_rate):
        return Assassin(name,100-failure_rate)
    @classmethod
    def failurePercentage(cls,name,failure_rate):
        return Assassin(name,100-failure_rate)

    def printDetails(self):
        print('Name:',self.name)
        print(f'Success rate: {self.success_rate}%')
        print('Total number of Assassin:',Assassin.no)



john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage('Akabane', 10)
akabane.printDetails()
```

```python
#Task 3


class Passenger:
    count=0
    def __init__(self,name):
        self.name=name
        self.total=450
        Passenger.count+=1
    def set_bag_weight(self,weight):
        self.weight=weight
        if self.weight<=20:
            self.total=self.total
        elif 21<=self.weight<=50:
            self.total+=50
        elif self.total>50:
            self.total+=100

    def printDetail(self):
        print(f'Name: {self.name}')
        print(f'Bus Fare: {self.total} taka')

print('Total Passenger:', Passenger.count)
p1 = Passenger('Jack')
p1.set_bag_weight(90)
p2 = Passenger('Carol')
p2.set_bag_weight(10)
p3 = Passenger('Mike')
p3.set_bag_weight(25)
print("=========================")
p1.printDetail()
print("=========================")
p2.printDetail()
print("=========================")
p3.printDetail()
print("=========================")
print('Total Passenger:', Passenger.count)
```

```python
#Task 5


class Employee:
    def __init__(self,name,workingPeriod):
        self.name=name
        self.workingPeriod=workingPeriod

    @classmethod
    def employeeByJoiningYear(cls,name,joining_year):
        return Employee(name,2021-joining_year)
    @staticmethod
    def experienceCheck(working_period,gender):
        if gender=="male":
            if working_period>3:
                return "He is experienced"
            else:
                return "He is not experienced"
        elif gender=="female":
            if working_period>3:
                return "She is experienced"
            else:
                return "She is not experienced"
employee1 = Employee('Dororo', 3)
employee2 = Employee.employeeByJoiningYear('Harry', 2016)
print(employee1.workingPeriod)
print(employee2.workingPeriod)
print(employee1.name)
print(employee2.name)
print(Employee.experienceCheck(2, "male"))
print(Employee.experienceCheck(3, "female"))
```

```python
#Task 6


class Laptop:
    laptopCount=0
    def __init__(self,name,count):
        self.name=name
        self.count=count
        Laptop.laptopCount+=count
    @staticmethod
    def advantage():
        print('Laptops are portable')
    @classmethod
    def resetCount(cls):
        Laptop.laptopCount=0

lenovo = Laptop("Lenovo", 5)
dell = Laptop("Dell", 7)
print(lenovo.name, lenovo.count)
print(dell.name, dell.count)
print("Total number of Laptops", Laptop.laptopCount)
Laptop.advantage()
Laptop.resetCount()
print("Total number of Laptops", Laptop.laptopCount)
```

```python
#Task 7


class Cat:
    Number_of_cats=0
    def __init__(self,name,work):
        self.name=name
        self.work=work
        Cat.Number_of_cats+=1
    @classmethod
    def no_parameter(cls):
        return cls(name='White',work='sitting')
    @classmethod
    def first_parameter(cls,name2):
        return cls(name2,work='sitting')
    @classmethod
    def second_parameter(cls,work2):
        return cls(name='Grey',work=work2)

    def changeColor(self,colour):
        self.name=colour
    def printCat(self):
        print(f'{self.name} cat is {self.work}')


print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
#Task 8


import math
class Cylinder:
    radius=5
    height=18
    def __init__(self,radius,height):
        self.radius=radius
        self.height=height
        print(f'Default radius={Cylinder.radius} and height={Cylinder.height}.')
        Cylinder.radius = self.radius
        Cylinder.height = self.height
        print(f'Updated radius={self.radius} and height={self.height}.')
    @classmethod
    def swap(cls,new_radius,new_height):
        return cls(new_height,new_radius)
    @classmethod
    def changeFormat(cls,st):
        radius,height=st.split('-')
        return cls(float(radius),float(height))
    @staticmethod
    def area(radius,height):
        #print(f'Area: {(2 * math.pi * Cylinder.radius * Cylinder.radius) + Cylinder.height}')
        print(f'Area: {(2*math.pi*Cylinder.radius*Cylinder.radius)+(2*math.pi*Cylinder.radius*Cylinder.height)}')
    @staticmethod
    def volume(radius,height):
        print(f'Volume: {math.pi*(Cylinder.radius)*Cylinder.radius*Cylinder.height}')
c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```

```python
#Task 10


class SultansDine:
    num_of_branch=0
    total_sell=0
    food_list=[]
    def __init__(self,branch_name):
        self.branch_name=branch_name
        SultansDine.num_of_branch+=1
        self.branch_sell=0

    def sellQuantity(self,quantity):
        if quantity < 10:
            self.branch_sell = quantity * 300
        elif quantity < 20:
            self.branch_sell = quantity * 350
        else:
            self.branch_sell = quantity * 400
        SultansDine.total_sell+=self.branch_sell
        SultansDine.food_list.append(self)
    def branchInformation(self):
        print(f'Branch Name: {self.branch_name}')
        print(f'Branch sell: {self.branch_sell} Taka')

    @classmethod
    def details(cls):
        print(f'Total Number of branch(s): {cls.num_of_branch}')
        print(f'Total sell: {cls.total_sell} Taka')
        for item in cls.food_list:
            print(f'Branch Name: {item.branch_name}, Branch Sell: {item.branch_sell} Taka')
            print(f'Branch consists of total sell\'s: {((item.branch_sell / cls.total_sell) * 100):.2f}%')


SultansDine.details()
print('########################')
dhanmodi = SultansDine('Dhanmondi')
dhanmodi.sellQuantity(25)
dhanmodi.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```
