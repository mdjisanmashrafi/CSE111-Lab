```python
#Task 3

def func(mi,mx,div):
    mi=int(mi)
    mx=int(mx)
    div=int(div)
    s=0
    for i in range(mi,mx):
        if i%div==0:
            s+=i
    return s
st = input()[1:-1].split(',')
print(func(st[0],st[1],st[2]))
```

```python
#Task 4

def func(snack,place='Mohakhali'):
    total=0
    if snack=='BBQ Chicken Cheese Burger':
        total+=250
    if snack=='Beef Burger':
        total+=170
    if snack=='Naga Drums':
        total+=200
    total=total+(total/100*8)
    if place=='Mohakhali':
        total+=40
    else:
        total+=60
    return total
st = input()[1:-1].split(',')
if len(st)==1:
    print(func(st[0][1:-1]))
else:
    print(func(st[0][1:-1],st[1][1:-1]))
```

```python
#Task5

def replace_domain(email, new_dom,old_dom=None):
    lis=email.split('@')
    if lis[1]==old_dom:
        print('Changed: '+lis[0]+'@'+new_dom)
    else:
        print('Unchanged: '+email)
st = input()[1:-1].split(', ')
if len(st)==2:
    replace_domain(st[0][1:-1],st[1][1:-1])
else:
    replace_domain(st[0][1:-1],st[1][1:-1],st[2][1:-1])

#(‘alice@kaaj.com’, ‘sheba.xyz’, ‘kaaj.com’)
#('alice@kaaj.com', 'sheba.xyz', 'kaaj.com')
#('bob@sheba.xyz', 'sheba.xyz')
```

```python
#Task 6

def vowel(string):
    word='aeiou'
    coun=''
    out=''
    lis=[]
    for i in string:
        if i in word:
            lis.append(i)
    if len(lis)==0:
        print('No vowels in the name')
    else:
        for j in lis:
            coun+=j
        for x in range(len(coun)-1):
            out+=coun[x]+','
        out+=coun[len(coun)-1]
        print('Vowels:',out,'.Total number of vowels:',len(lis))
st=input()[1:-1]
vowel(st)
```

```python
#Task 7

def func(word):
    string = word[::-1]
    if word==string:
        return 'Palindrome'
    else:
        return 'Not a palindrome'

st = input()[1:-1]
new=''
for i in st:
    if i==' ':
        pass
    else:
        new+=i
print(func(new))
```

```python
#Task 9


def punc(sentence):
    flag=True  #cap
    out=''
    for i in range(len(sentence)):
        if flag==True:
            if sentence[i].isalpha()==True:
                out+=sentence[i].upper()
                flag=False
            else:
                out+=sentence[i]
        elif sentence[i] in '.!?' :
            out+=sentence[i]
            flag=True
        elif sentence[i]=='i':
            if sentence[i-1]==' ' and sentence[i+1] in '.?!':
                out+=sentence[i].upper()
            else:
                out+=sentence[i]
        else:
            out+=sentence[i]
    return out
st=input()[1:-1]
print(punc(st))
```
