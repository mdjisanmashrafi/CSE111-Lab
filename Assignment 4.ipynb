```python
#Task 1



class Calculator:
    def __init__(self):
        print('Let''\'s Calculate')
    def add(self,value1,value2):
        total=value1+value2
        return total
    def substract(self, value1, value2):
        total=value1-value2
        return total
    def multiply(self, value1, value2):
        total = value1 * value2
        return total
    def devide(self, value1, value2):
        total=value1/value2
        return total
cal=Calculator()
val1=int(input('Value 1 : '))
operator=input('Operator : ')
val2=int(input('Value 2 : '))
if operator=='+':
    print('Result: ',cal.add(val1,val2))
if operator=='-':
    print('Result: ',cal.substract(val1,val2))
if operator=='*':
    print('Result: ',cal.multiply(val1,val2))
if operator=='/':
    print('Result: ',cal.devide(val1,val2))
```

```python
#Task 21


class Hospital:
    def __init__(self,hospital_name):
        self.doctor=[]
        self.patient=[]
        self.doctor_dc={}
        self.patient_dc={}
        self.doctor_count=0
        self.patient_count=0
        self.hospital_name=hospital_name

    def addDoctor(self,d):
        self.doctor.append(d)
        self.doctor_dc[d.doc_id] = [d.doc_name,d.speaciality]
        self.doctor_count += 1
    def addPatient(self,p):
        self.patient.append(p)
        self.patient_dc[p.patient_id]=[p.patient_name,p.age,p.phn_no]
        self.patient_count += 1


    def getDoctorByID(self,doc_id):
        for i in self.doctor:
            if i.doc_id==doc_id:
                dc='Doctor\'s ID: '+i.doc_id+'\n'+'Name: '+i.doc_name+'\n'+'Speciality: '+i.speaciality
                return dc

    def getPatientByID(self,patient_id):
        for i in self.patient:
            if i.patient_id == patient_id:
                pa = 'Patient\'s ID: ' + i.patient_id + '\n' + 'Name: ' + i.patient_name + '\n' + 'Age: ' + str(i.age)+'\n'+'Phn no.:'+str(i.phn_no)
                return pa

    def allDoctors(self):
        print('All Doctors:')
        print('Number of Doctors:',self.doctor_count)
        print(self.doctor_dc)

    def allPatients(self):
        print('All Patients:')
        print('Number of Doctors:',self.patient_count)
        print(self.patient_dc)

class Doctor:
    def __init__(self,doc_id,profession,doc_name,speaciality):
        self.doc_id=doc_id
        self.profession=profession
        self.doc_name=doc_name
        self.speaciality=speaciality


class Patient:
    def __init__(self,patient_id,profession,patient_name,age,phn_no):
        self.patient_id=patient_id
        self.profession=profession
        self.patient_name=patient_name
        self.age=age
        self.phn_no=phn_no




h = Hospital("Evercare")
d1 = Doctor("1d","Doctor", "Samar Kumar", "Neurologist")
h.addDoctor(d1)
print("=================================")
print(h.getDoctorByID("1d"))
print("=================================")
p1 = Patient("1p","Patient", "Kashem Ahmed", 35, 12345)
h.addPatient(p1)
print("=================================")
print(h.getPatientByID("1p"))
print("=================================")
p2 = Patient ("2p","Patient", "Tanina Haque", 26, 33456)
h.addPatient(p2)
print("=================================")
print(h.getPatientByID("2p"))
print("=================================")
h.allDoctors()
h.allPatients()
```
