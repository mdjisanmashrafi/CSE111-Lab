```python
#Task 10


class Student:
    def __init__(self,name,ID):
        self.name = name
        self.ID = ID
    def Details(self):
        return "Name: "+self.name+"\n"+"ID: "+self.ID+"\n"


class CSEStudent(Student):
    def __init__(self,name,ID,semester):
        super().__init__(name,ID)
        self.semester=semester
        self.dc={}

    def addCourseWithMarks(self,*results):
        self.results=results
        for i in range(0,len(results),2):
            self.dc[results[i]]=results[i+1]
    def Details(self):
        x=super().Details()
        return x+'Current Semester: '+self.semester
    def showGPA(self):
        self.cg=0
        self.cg_lis=[]
        self.course_count=[]

        print(f'Bob has taken {int(len(self.results)/2)} courses.')
        for name,mark in self.dc.items():
            self.course_count.append(1)
            if mark>=85:
                self.cg = 4.0
                self.cg_lis.append(self.cg)

            elif 80<=mark<=84:
                self.cg = 3.3
                self.cg_lis.append(self.cg)

            elif 70<=mark<=79:
                self.cg = 3.0
                self.cg_lis.append(self.cg)

            elif 65<=mark<=69:
                self.cg = 2.3
                self.cg_lis.append(self.cg)

            elif 57<=mark<=64:
                self.cg = 2.0
                self.cg_lis.append(self.cg)

            elif 55<=mark<=56:
                self.cg = 1.3
                self.cg_lis.append(self.cg)

            print(f'{name}: {self.cg}')
        self.gpa = sum(self.cg_lis*3)/sum(self.course_count*3)
        print(f'GPA of {self.name} is: {self.gpa:.2f}')



Bob = CSEStudent("Bob","20301018",'Fall 2020')
Carol = CSEStudent("Carol","16301814",'Fall 2020')
Anny = CSEStudent("Anny","18201234",'Fall 2020')
print("#########################")
print(Bob.Details())
print("#########################")
print(Carol.Details())
print("#########################")
print(Anny.Details())
print("#########################")
Bob.addCourseWithMarks("CSE111",83.5,"CSE230",73.0,"CSE260",92.5)
Carol.addCourseWithMarks("CSE470",62.5,"CSE422",69.0,"CSE460",76.5,"CSE461",87.0)
Anny.addCourseWithMarks("CSE340",45.5,"CSE321",95.0,"CSE370",91.0)
print("----------------------------")
Bob.showGPA()
print("----------------------------")
Carol.showGPA()
print("----------------------------")
Anny.showGPA()
```

```python
class Author:
    nov_c=0
    rom_c=0
    def __init__(self,name):
        self.name=name
        self.nov=[]
        self.rom=[]
        self.lis=[]
    def addBook(self,*b_lis):
        self.b_lis=b_lis
        for b in b_lis:
            if b.name==None:
                print('A book canâ€™t be added without name')
            else:
                print(f'A book named {b.name} is added')
            if b.type=='Novel' and b.name!=None:
                Author.nov_c += 1
                self.nov.append(b.name)
            if b.type=='Romantic':
                Author.rom_c+=1
                self.rom.append(b.name)
            self.lis.append((b.name,b.type))
    def showAllBooks(self):
        print(f'Author Name: {self.name}')
        print(f'Novel: {self.nov}')
        print(f'Romantic: {self.rom}')
    def removeLastBook(self):
        if self.lis[len(self.lis)-1][1]=='Romantic':
            Author.rom_c-=1
        else:
            Author.nov_c-=1
        self.lis.pop(len(self.lis)-1)
    @staticmethod
    def showAllInfo():
        print(f'Total Novel books: {Author.nov_c}')
        print(f'Total Romantic books: {Author.rom_c}')

class Book:
    def __init__(self,*lis):
        if len(lis)==1:
            self.name=None
            self.type=lis[0]
        if len(lis)==2:
            self.name=lis[0]
            self.type=lis[1]
    def setName(self,new):
        self.name=new


a1 = Author("David")
b1 = Book("Novel")
b2 = Book("The Notebook", "Romantic")
b3 = Book("The Alchemist", "Novel")
print("==============================")
a1.addBook(b1)
print("==============================")
b1.setName("The Kite runner")
a1.addBook(b1)
print("==============================")
a1.addBook(b2, b3)
print("==============================")
a1.addBook(Book("The Fault in Our Stars", "Romantic"))
print("==============================")
a1.showAllBooks()
print("==============================")
a1.removeLastBook()
print("==============================")
a1.showAllBooks()
print("==============================")
Author.showAllInfo()
```

```python
class Club:
   def __init__(self, name, league):
     self.name = name
     self.league = league

   def __str__(self):
     s = f"Name: {self.name}\nLeague: {self.league}"
     return s
class Madrid(Club):
    def __init__(self,name, league,coun):
        super().__init__(name, league)
        self.country=coun
        self.dc={}
    def addTrophy(self,*trophy):
        self.trophy=trophy
        for i in self.trophy:
            x,y=i.split(' ')
            if x not in self.dc.keys():
                self.dc[x]=[y]
            else:
                self.dc[x].append(y)

    def showClubPoint(self):
        print(f'{self.name} has {len(self.trophy)*10} points')

    def __str__(self):
        x = super().__str__() + '\nCountry: ' + self.country + '\nTrophy Cabinet: \n'
        for k, v in self.dc.items():
            x += k + ':' + str(v) + '\n'

        return x

class Barca(Club):
    def __init__(self,name, league,coun):
        super().__init__(name, league)
        self.country=coun
        self.dc = {}
    def addTrophy(self, *trophy):
        self.trophy = trophy
        for i in self.trophy:
            x, y = i.split(' ')
            if x not in self.dc.keys():
                self.dc[x] = [y]
            else:
                self.dc[x].append(y)

    def showClubPoint(self):
        print(f'{self.name} has {len(self.trophy)*10} points')

    def __str__(self):
        x=super().__str__()+'\nCountry: '+self.country+'\nTrophy Cabinet: \n'
        for k,v in self.dc.items():
            x+=k+':'+str(v)+'\n'

        return x


# Write your codes here.
# Do not change the following lines of code.
c1 = Madrid("Real Madrid", "La Liga", "Spain")
c1.addTrophy("LaLiga 2020","UEFA 2016","UEFA 2017","LaLiga 2019","CopaDelRay 2014")
print('====================================')
c1.showClubPoint()
print('====================================')
print(c1)
print('====================================')
c2 = Barca("Barcelona", "La Liga", "Spain")
c2.addTrophy("LaLiga 2015","UEFA 2015", "LaLiga 2016", "UEFA 2017", "LaLiga 2020", "Supercopa 2020")
print('====================================')
c2.showClubPoint()
print('====================================')
print(c2)
```
