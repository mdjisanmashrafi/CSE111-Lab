```python
#Task 1

dc={'Fname':[],'Mname':[],'Lname':[]}
n=int(input())
for i in range(n):
    name=input().split()
    if len(name)==1:
        dc['Fname'].append(name[0])
    elif len(name)==2:
        dc['Fname'].append(name[0])
        dc['Lname'].append(name[1])
    elif len(name)==3:
        dc['Fname'].append(name[0])
        dc['Mname'].append(name[1])
        dc['Lname'].append(name[2])
print(dc)
```

```python
#Task 4

dc={}
st='AABAABBAA'
for i in range(1,len(st)//2+1):
    x='A'*i+'B'*i
    y='B'*i+'A'*i
    if x in st:
        dc[x]=st.count(x)
    if y in st:
        dc[y]=st.count(y)
mx=''
for k,v in dc.items():
    print(k+'-'+str(v)+' times')
for i in dc.keys():
    if len(i)>len(mx):
        mx=i
print('Longest pattern',mx)
print(dc)
```
