```python
#Task 1


dc1={'a': 100, 'b': 100, 'c': 200, 'd': 300}
dc2={'a': 300, 'b': 200, 'd': 400, 'e': 200}
out_dc={}
for k1,v1 in dc1.items():
    if k1 in dc2.keys():
        out_dc[k1]=v1+dc2[k1]
    else:
        out_dc[k1]=v1
for k2,v2 in dc2.items():
    if k2 not in out_dc:
        out_dc[k2]=v2
values=[]
for i in out_dc.values():
    if i not in values:
        values.append(i)
values.sort()
print(out_dc)
print('Values:',tuple(values))
```
